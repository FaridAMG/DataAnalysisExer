```
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

pi = np.pi
# Needed Variables
I1=50e-3
I2=50e-3
# I1I2=2526.3125
# IAmp=2083.125
VpW=1250           # Volts per Watt
mMax=1.2           # Modulation depth m*pi
mMin=0.8
mTyp=0.93
c =299792458       # Speed of Light
e =1.6022e-019     # Electron Charge
eo =8.8542e-012    # Epsilon zero
me =9.1094e-031    # Electron Mass
Wl = 10.6e-6       # Wavelenght
Wm=2*pi*50000      # Modulation Frequency

spp=200
numP=2#20
Nsamples = numP*spp
#t=(0:Nsamples-1)*1e-7
t=np.linspace(0, Nsamples*1e-7, Nsamples)
PH= 1.5 # Test phase value
resPhi = 5000



#%%
Ur=(I1+I2+2*np.sqrt(I1*I2)*np.cos(pi*mTyp*np.sin(Wm*t)+PH))
Amp=max(Ur)-min(Ur)
noiseLevel=0.1
noise=np.random.rand(1,t.size)*Amp*noiseLevel
UrNsy=Ur+noise
UrNsy = UrNsy.reshape(1,numP,spp)

print(UrNsy)

plt.plot(UrNsy[0,0])





```
